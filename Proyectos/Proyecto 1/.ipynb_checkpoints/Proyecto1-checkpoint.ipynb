```python
# TICTACTOE
import random


class Player:
    def __init__(self,player_number):
        self.name = ""
        self.token = ""
        self.firstTurn = None
        self.player_number = player_number

    def __str__(self):
        player_info = "\nNombre: {} Token: {}. Es tu turno\n".format(self.name, self.token)
        return player_info

    def set_name(self):
        while True:
            name = input('Jugador {} ingrese su nombre: '.format(self.player_number))
            if name == 'COM':
                print('Por favor, elija otro nombre')
            else:
                self.name = name
                break

    def set_token(self):
        while True:
            token = input(self.name + ' desea jugar como X o O? ').upper()
            if token not in ('X', 'O'):
                print("Token invalido")
            else:
                self.token = token
                break

    def set_first_turn(self, value):
        self.firstTurn = value

    def reverse_turn(self):
        if self.firstTurn:
            self.firstTurn = False
        else:
            self.firstTurn = True

    def winner(self):
        winner = "\nNombre: {} Token: {}. Has ganado!\n".format(self.name, self.token)
        return winner


def draw_board():
    board = [None] + list(range(1, 10))
    print(str(board[7]) + '|' + str(board[8]) + '|' + str(board[9]) + '\n' + '-|-|-\n' + \
          str(board[4]) + '|' + str(board[5]) + '|' + str(board[6]) + '\n' + '-|-|-\n' + \
          str(board[1]) + '|' + str(board[2]) + '|' + str(board[3])
          )
    return board


def update_board(board):
    print(str(board[7]) + '|' + str(board[8]) + '|' + str(board[9]) + '\n' + '-|-|-\n' + \
          str(board[4]) + '|' + str(board[5]) + '|' + str(board[6]) + '\n' + '-|-|-\n' + \
          str(board[1]) + '|' + str(board[2]) + '|' + str(board[3])
          )
    updated_board = board
    return updated_board


def number_players():
    while True:
        number_of_players = input("Partida de uno o dos jugadores? ")
        if number_of_players not in ('1', '2'):
            print("Numero de jugadores invalido")
        else:
            break
    return number_of_players


def choose_reverse_token(player):
    if player.token == 'X':
        token = 'O'
    elif player.token == 'O':
        token = 'X'
    return token


def create_com_player(p1):
    com_player = Player(2)
    com_player.name = "COM"
    com_player.token = choose_reverse_token(p1)
    return com_player


def choose_first_turn(p1, p2):
    player_list = []
    player_list.append(p1)
    player_list.append(p2)
    first_player = random.choice(player_list)
    if first_player == p1:
        p1.set_first_turn(True)
        p2.set_first_turn(False)
    elif first_player == p2:
        p1.set_first_turn(False)
        p2.set_first_turn(True)
    return p1, p2


def create_players(num_players):
    if num_players == '1':
        p1_temp = Player(1)
        p1_temp.set_name()
        p1_temp.set_token()
        p2_temp = create_com_player(p1_temp)
        p1, p2 = choose_first_turn(p1_temp, p2_temp)
    elif num_players == '2':
        p1_temp = Player(1)
        p2_temp = Player(2)
        p1_temp.set_name()
        p1_temp.set_token()
        p2_temp.set_name()
        p2_temp.token = choose_reverse_token(p1_temp)
        p1, p2 = choose_first_turn(p1_temp, p2_temp)
    return p1, p2


def choose_position(player, board):
    while True:
        if player.name == 'COM':
            a = random.randint(0, 9)
        else:
            a = int(input())
        if a in board:
            board[a] = player.token
            break
        else:
            print("\nMovimiento Invalido")
    return update_board(board)


def game_over(board):
    if board[1] == board[2] == board[3]:
        return True
    elif board[4] == board[5] == board[6]:
        return True
    elif board[7] == board[8] == board[9]:
        return True
    elif board[1] == board[4] == board[7]:
        return True
    elif board[2] == board[5] == board[8]:
        return True
    elif board[3] == board[6] == board[9]:
        return True
    elif board[1] == board[5] == board[9]:
        return True
    elif board[3] == board[5] == board[7]:
        return True
    else:
        return False


def game(board, playerList):
    #gameover = None
    updated_board = board
    active_player = None
    while not game_over(updated_board):
        for player in playerList:
            if player.firstTurn:
                active_player = player
                print(player)
                updated_board = choose_position(player, board)
            player.reverse_turn()
    print("Felicidades {}, has ganado!".format(active_player.name))

def main():
    n = number_players()
    pl = create_players(n)
    b = draw_board()
    game(b, pl)


if __name__ == "__main__":
    main()

```
