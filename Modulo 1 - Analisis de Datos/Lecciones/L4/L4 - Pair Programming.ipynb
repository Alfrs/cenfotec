```python
from random import randint
print(r"Bienvenido al juego")
print(r"Voy a pensar en un numero del 1 al 100, y vas a adivinarlo")

system_num = randint(1,100)
tries_num = 0
check_num = 0
prev_pred_num = None
min_num = 1
max_num = 100

print(system_num)
def new_range(user_num):
    if(user_num > system_num):
        max_num = user_num
        return max_num
    elif(user_num < system_num):
        min_num = user_num
        return min_num

while True:
    #user_num = int(input(r"Introduzca un numero: "))
    user_num = randint(min_num,max_num)
    tries_num += 1
    
    if user_num > 100 or user_num < 1:
        print("Debes introducir un numero entre 1 y 100")
        continue
    if user_num == system_num:
        print(r"Felicidades! Has adivinado que el numero era {}! Solo tomo {} intentos".format(system_num,tries_num))
        break
    if prev_pred_num:
        if abs(system_num - user_num) < abs(system_num - prev_pred_num):
            print(r"Te estas acercando! Tu numero:", str(user_num))
            prev_pred_num = user_num
            new_range(user_num)
        else:
            print(r"Te estas alejando :( Tu numero:",str(user_num))
            prev_pred_num = user_num
            new_range(user_num)
    else:
        if abs(system_num - user_num) <= 10:
            print("Uy! En el poste, casi casi. Tu numero:",str(user_num))
            prev_pred_num = user_num
            new_range(user_num)
        else:
            print("Que va, muy lejos. Tu numero:",str(user_num))
            prev_pred_num = user_num
            new_range(user_num)
    
    
```
